# Task 22

Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean. Plot
it on a histogram and comment on its distribution.

```python
mean = []

for i in range(100):
  
  for x in range(0,1001,50):
    
    mean1 = data_file['damageDealt'].head(x).mean()
    mean.append(mean1)
 
plt.hist(data = data_file['damageDealt'], x = mean);
```
